# DS Case Study (Hackathon)

# Case Study Data

```python
import pandas as pd
import os

df = pd.read_csv(os.path.join(os.getcwd(), 'case_study_data.csv'))
cat_features = ['cat_feature_1', 'cat_feature_2']
df = pd.get_dummies(df, columns=cat_features, drop_first=True)
df
```

# Data Prep

## Data Separation (X and Y)

```python
y = df['target']
y
```

```python
x = df.drop('target', axis=1)
x
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
x_train
```

```python
x_test
```

# Model Building

## Linear Regression

### Training the Model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

### Applying the Model for Predictions

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

lr_train_sq = mean_squared_error(y_train, y_lr_train_pred)
lr_train_abs = mean_absolute_error(y_train, y_lr_train_pred)

lr_test_sq = mean_squared_error(y_test, y_lr_test_pred)
lr_test_abs = mean_absolute_error(y_test, y_lr_test_pred)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_sq, lr_train_abs, lr_test_sq, lr_test_abs]).transpose()
lr_results.columns = ['Method', 'Training SQ', 'Training ABS', 'Test SQ', 'Test ABS']
```

```python
lr_results
```

## Random Forest

### Training the Model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=50)
rf.fit(x_train, y_train)
```

### Applying the Model for Prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

rf_train_sq = mean_squared_error(y_train, y_rf_train_pred)
rf_train_abs = mean_absolute_error(y_train, y_rf_train_pred)

rf_test_sq = mean_squared_error(y_test, y_rf_test_pred)
rf_test_abs = mean_absolute_error(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_sq, rf_train_abs, rf_test_sq, rf_test_abs]).transpose()
rf_results.columns = ['Method', 'Training SQ', 'Training ABS', 'Test SQ', 'Test ABS']
rf_results
```

## Decision Tree

### Training the Model

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(max_depth=2, random_state=50)
dt.fit(x_train, y_train)
```

### Applying the Model for Predictions

```python
y_dt_train_pred = dt.predict(x_train)
y_dt_test_pred = dt.predict(x_test)
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

dt_train_sq = mean_squared_error(y_train, y_dt_train_pred)
dt_train_abs = mean_absolute_error(y_train, y_dt_train_pred)

dt_test_sq = mean_squared_error(y_test, y_dt_test_pred)
dt_test_abs = mean_absolute_error(y_test, y_dt_test_pred)
```

```python
dt_results = pd.DataFrame(['Decision Tree', dt_train_sq, dt_train_abs, dt_test_sq, dt_test_abs]).transpose()
dt_results.columns = ['Method', 'Training SQ', 'Training ABS', 'Test SQ', 'Test ABS']
dt_results
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results, dt_results], axis=0).reset_index(drop=True)
df_models
```
